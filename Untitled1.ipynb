```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from xgboost import XGBClassifier

# 1. LOAD DATA
df = pd.read_csv("telco.csv")

# 2. SELECT FEATURES
df_model = df[[
    "customerID",
    "tenure",
    "MonthlyCharges",
    "TotalCharges",
    "Contract",
    "InternetService",
    "PaymentMethod",
    "Churn"
]].copy()

# 3. CLEAN TotalCharges
df_model["TotalCharges"] = pd.to_numeric(df_model["TotalCharges"], errors="coerce")
df_model["TotalCharges"] = df_model["TotalCharges"].fillna(df_model["TotalCharges"].median())

# 4. ENSURE CHURN IS CLEAN (NO MAPPING NEEDED)
df_model["Churn"] = df_model["Churn"].astype(int)

# 5. ENCODE CATEGORICAL FEATURES
le = LabelEncoder()

for col in ["Contract", "InternetService", "PaymentMethod"]:
    df_model[col] = le.fit_transform(df_model[col].astype(str))

# 6. FINAL CLEANUP
df_model = df_model.replace([np.inf, -np.inf], np.nan)
df_model = df_model.fillna(0)

# 7. FEATURES & TARGET
X = df_model.drop(["Churn", "customerID"], axis=1)
y = df_model["Churn"]

# 8. CHECK CLASS BALANCE
print("Churn distribution:\n", y.value_counts())

# 9. TRAIN TEST SPLIT (IMPORTANT: STRATIFY)
X_train, X_test, y_train, y_test = train_test_split(
    X.values,
    y.values,
    test_size=0.2,
    random_state=42,
    stratify=y
)

# 10. MODEL
model = XGBClassifier(
    random_state=42,
    eval_metric="logloss",
    objective="binary:logistic"
)

# 11. TRAIN
model.fit(X_train, y_train)

# 12. PREDICTIONS
df_model["Churn_Probability"] = model.predict_proba(X.values)[:, 1]
df_model["Predicted_Churn"] = model.predict(X.values)

# 13. RISK LEVEL
df_model["Risk_Level"] = df_model["Churn_Probability"].apply(
    lambda x: "High Risk" if x > 0.7 else "Medium Risk" if x > 0.4 else "Low Risk"
)

# 14. FEATURE IMPORTANCE
importance = pd.DataFrame({
    "Feature": X.columns,
    "Importance": model.feature_importances_
}).sort_values(by="Importance", ascending=False)

# 15. EXPORT FOR POWER BI
df_model.to_csv("telco_ml_output.csv", index=False)
importance.to_csv("feature_importance.csv", index=False)

print("MODEL SUCCESSFULLY TRAINED")
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score

# 1. PREDICTIONS ON TEST SET
y_pred = model.predict(X_test)
y_prob = model.predict_proba(X_test)[:, 1]

# 2. ACCURACY
print("Accuracy:", accuracy_score(y_test, y_pred))

# 3. CONFUSION MATRIX
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))

# 4. CLASSIFICATION REPORT
print("Classification Report:\n", classification_report(y_test, y_pred))

# 5. AUC SCORE (VERY IMPORTANT FOR INTERVIEWS)
print("ROC-AUC Score:", roc_auc_score(y_test, y_prob))
```
